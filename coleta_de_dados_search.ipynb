```python
import pandas as pd
import numpy as np
import re
import time
import json

import requests as rq
import bs4 as bs4
```

```python
queries = ["machine+learning", "data+science", "kagle"]
url = "https://youtube.com/results?search_query={query}&sp=CAI%253D&p={page}"
```

# Coletando os Dados

```python
for query in queries:
    for page in range(1,50):
        urll = url.format(query=query, page=page) #substituindo os parametros
        print(urll)
        response = rq.get(urll) #realizando a requisição ao site definido
        
        #Esquevendo o arquivo com os dados obtidos do site
        with open("./dados_brutos/{}_{}.html".format(query, page), 'w+', encoding="utf-8") as output:
            output.write(response.text)
            
        #time.sleep(2)
```

# Processamento dos dados Brutos

```python
for query in queries:
    for page in range(1,21):
        with open("./dados_brutos/{}_{}.html".format(query, page), 'r+', encoding="utf-8") as inp:
            page_html = inp.read() #Lendo dados dos arquivos
            
            parsed = bs4.BeautifulSoup(page_html)
            
            tags = parsed.findAll("a") #Retorna uma lista de todas as tags "a"
            
            for e in tags: 
                if e.has_attr("aria-describedby"): #Verifica se existe o attributo dentro da lista
                    link = e['href']
                    title = e['title']
                    with open("parsed_videos.json", 'a+') as output:
                        data = {"link": link, "title": title, "query": query}
                        output.write("{}\n".format(json.dumps(data))) #Transformando em json
                        
# O parametro 'a+' está adicionando novas linhas (append)
```
